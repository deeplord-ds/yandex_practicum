# Описание процесса обработки стали

На металлургическом комбинате  сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша. 

Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»).

Перед тем, как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.

Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. *slab*, «плита»).

**Влияние параметров технологического процесса на предсказания модели**

При использовании модели и планировании параметров технологического процесса необходимо использовать следующие закономерности:
   - Три главных признака (начальная температура, средняя мощность и кол-во нагревов) прямо коррелируют с конечной температурой. Чем они больше, тем выше будет предсказанная моделью температура и тем сильнееданный параметр будет влиять на предсказание. Обратное тоже верно.
   - Увеличение объема проволочныого материала Wire 1 увеличит конечную температуру. Зависимость средняя по величине.
   - Увеличение объема проволочныого материала Wire 2 меньшит конечную температуру. Зависимость средняя по величине.
   - Добавление супучего материал bulk 15 увеличивает конечную температуру. Зависимость средняя по величине.
   - При росте объема сыпучих материалов bulk 12 и bulk 14 и bulk 4 конечная температура растет
   - При росте объема сыпучих материалов bulk 6 и bulk 1 и bulk 11 и bulk 3 конечная температура падает
   - Время процесса увеличивает конечную температуру, как  иуровень подачи газа. Степень влияния - слабая. 
   - Остальные признаки практически не влияют на конечную темпераутру.

Особенностью технологического процесса является то, что сильные факторы компенсируются только сильными факторами или коммулятивнм действием нескольких слабых. При этом не всегда возможно свободно менять каждый из них в отдельности (напр., сыпучие и проволочные добавляют исходя из доведения состава до нужного состояния, газ нельзя произвольно убирать в силу его влияяния на перемешивание и  т.д.). Однако технолог может составить с помощью модели карты процесса для отдельных базовых комбиаций параметров. Также можно попробовать составить такую карту, которая не будет противоречить логике технологии, но в тоже время уменьшит среднюю потребляемую мощность. Таким образом, модель может быть использована для исключения  заведомо невыполнимых комбинаций (невозможность компенсировать доминирующий признак другими с малым влиянием).<br>
